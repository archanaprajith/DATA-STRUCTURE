```python
#ENCODING THE SEX COLUMN
insurance_dataset.replace({'sex':{'male':0,'female':1}},inplace=True)
```

```python
#ENCOUDING THE SMOKER COLUMN
insurance_dataset.replace({'smoker':{'yes':0,'no':1}}, inplace=True)
```

```python
#ENCOUDING THE REGION COLUMN
insurance_dataset.replace({'regioner':{'southeast':0,'southwest':1,'northeast':2,'northwest':3}}, inplace=True)
```

```python
x=insurance_dataset.drop(columns='charges',axis=1)
y=insurance_dataset['charges']
```

```python
print(x)
```

```python
print(y)
```

```python
x_train, x_test, y_train, ytest=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

```python
#loading the linear regression model
regressor=LinearRegression()
```

```python
regressor.fit(x_train,y_train)
```

```python
#predining data
training_data_prediction=regressor.predict(x_train)
```

```python
r2_train=metrics.r2_score(y_train,training_data_prediction)
print('R squared value:', r2_train)
```

```python
#prediction on testing data
test_data_prediction=regressor.predict(x_test)
r2_test=metrics.r2_score(ytest,test_data_prediction)
print('R squared value:', r2_test)
```

```python
# Example input data: (age, sex, bmi, children, smoker, region_northwest, region_southeast, region_southwest)
# Note: The region values should be 0 or 1 based on the one-hot encoding
# For example, if the region is 'southwest', region_southwest should be 1, and others 0.
input_data=(31, 1, 25, 74, 0, 0, 0, 1) # Example for southwest region

# Create a DataFrame with the same column names as x_train
input_data_as_dataframe = pd.DataFrame([input_data], columns=x_train.columns)

prediction = regressor.predict(input_data_as_dataframe)
print(prediction)
print('The insurance is USD', prediction[0])
```
